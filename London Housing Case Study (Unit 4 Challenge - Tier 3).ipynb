```python
# [STEP 1] : find the avg house price in each borough in Sept 2001 and Sept 2021

# (1) split df into smaller df for each borough
# (2) extract avg_price for each df (i.e. borough) in 2001-09-01 and 2021-09-01
```

```python
df_grouped[df_grouped["Borough"]=="Westminster"][df_grouped["year"]==1995]["avg_price"]
```

```python
def price_change(df, borough, start_year, end_year):
    """
    Calculates the price change for "borough" from "start_year" to "end_year" by dividing "end_year" by "start_year"
    for a particular London borough.

    Example 1:
    The City of London had an average housing price of $100 in 2000 and $200 in 2020. 
    The price change would be $200/$100 = 2.
    This means that the average housing prices in the City of London increased by a factor of 2 from 2000 to 2020.

    Example 2:
    Camden had an average housing price of $200 in 2000 and $50 in 2020.
    The price change would be $50/$200 = 0.25.
    This means that the average housing prices in Camden increased by a factor of 0.25 from 2000 to 2020;
    or, in other words, the average housing prices decreased by a factor of 4 from 2000 to 2020.
    """  

    start_price = float(df[df["Borough"]==borough][df["year"]==start_year]["avg_price"].values)
    end_price = float(df[df["Borough"]==borough][df["year"]==end_year]["avg_price"].values)
    change = end_price / start_price
    
    return change
```

```python
price_change(df_grouped, "Westminster", 2000, 2020)
```

```python
def price_changes(df, start_year, end_year):
    price_changes = {}
    boroughs = list(df["Borough"].unique())
    for borough in boroughs:
        price_changes[borough] = [price_change(df, borough, start_year, end_year)]
        
    return price_changes
```

```python
def max_house_price_increase(df, start_year, end_year):
    price_changes_dict = price_changes(df, start_year, end_year)
    max_borough = max(price_changes_dict, key=price_changes_dict.get)
    
    return max_borough
```

```python
max_house_price_increase(df_grouped, 2000, 2020)
```

```python
price_changes(df_grouped, 2000, 2020)
```

```python
df_price_changes = pd.DataFrame(price_changes(df_grouped, 2000, 2020))
df_price_changes
```
